```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
class Vector(object):
    ''' a class of Vectors '''
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z

        self.length = (x**2 + y**2 + z**2)**0.5
        self.length_xy = (x**2 + y**2)**0.5
        # self.sin_angle_xy = self.z/self.length
        #
        #
        self.cos_angle_xy = self.length_xy/self.length
        # cos is in [0,1]

        self.angle_xy = np.arcsin(z/self.length)
        # sin is in [-1, 1]
        # angle in unit “radian”
        # angle is in [-pi, pi]

    def get_normal_vector(self):
        ''' get the normal vector defined by self and z axis '''
        normal_vector = Vector(-self.y/self.length_xy, self.x/self.length_xy, 0)
        return normal_vector
    
    def get_add(self, other):
        add = Vector(self.x + other.x, self.y + other.y, self.z + other.z)
        return add

    def get_subtract(self, other):
        subtract = Vector(self.x - other.x, self.y - other.y, self.z - other.z)
        return subtract

    def get_dot_product(self,other):
        dot_product = self.x * other.x + self.y * other.y + self.z * other.z
        return dot_product

    def get_projection_xy(self):
        projection_xy = Vector(self.x/self.cos_angle_xy, self.y/self.cos_angle_xy, 0)
        return projection_xy

    def get_projection(self, other):
        ''' get the projection of [self] on [other] '''
        D = self.get_dot_product(other)
        projection = Vector(D*other.x, D*other.y, D*other.z)
        return projection

    def get_rotated(self, alpha):
        ''' rotate alpha angle towards -z direction along '''
        ''' the plane defined by self and z_axis '''

        theta = self.angle_xy

        phi = theta - alpha

        m = np.cos(phi)/np.cos(theta)
        # m is the scaling factor of x and y

        n = np.sin(phi)/np.sin(theta)
        # n is the scaling factor of z

        rotated = Vector(self.x * m, self.y * m, self.z * n)
        return rotated

    def get_induced_angle(self, other):
        D = self.get_dot_product(other)
        cos_angle = D/(self.length * other.length)
        induced_angle = np.arccos(cos_angle)
        return induced_angle
    
    def to_list(self):
        vec = [0, 0, 0, 0, 0, 0]
        vec[3] = self.x
        vec[4] = self.y
        vec[5] = self.z
        return vec
```

```python
# Vector AB [-2.43640845  0.32606877  1.34834414]
# Vector AC [-0.65555028 -1.13059794  0.56823733]


AB1 = Vector(-2.43640845, 0.32606877, 1.34834414)
print('AB1.length', AB1.length)
alpha = AB1.angle_xy

Nv = AB1.get_normal_vector()


twist_angle = np.degrees(alpha)

print('twist_angle', twist_angle)

AB2 = AB1.get_projection_xy()

AC1 = Vector(-0.65555028, -1.13059794,  0.56823733)




AC_Nv = AC1.get_projection(Nv)
# this vector is constant during the rotation of AB to xy plane

AC1_p1 = AC1.get_subtract(AC_Nv)

AC1_p2 = AC1_p1.get_rotated(alpha)

AC2 = AC1_p2.get_add(AC_Nv)

AD = AC2.get_projection(AB2)

DC = AC2.get_subtract(AD)

rotation_angle = DC.get_induced_angle(AC_Nv)

rotation_angle  = np.degrees(rotation_angle)
print('rotation_angle', rotation_angle)
```
